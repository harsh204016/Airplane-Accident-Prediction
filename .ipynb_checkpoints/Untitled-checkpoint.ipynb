```python
from sklearn import preprocessing
lbl=preprocessing.LabelEncoder()
lbl.fit(train.iloc[:,0])
train.iloc[:,0]=lbl.transform(train.iloc[:,0])
```

```python
X=train.drop(['Severity'],axis=1)
Y=train['Severity']
X.shape,Y.shape
```

```python
Y.value_counts()
```

```python
%%time
from sklearn.metrics import confusion_matrix,accuracy_score
from lightgbm import LGBMClassifier
from sklearn.model_selection import StratifiedKFold

lgbm = LGBMClassifier(objective='multiclass', random_state=5,n_jobs=-1,learning_rate=0.2,num_leaves= 10,
min_data=50,max_depth=30)
skf = StratifiedKFold(n_splits=10,shuffle=True)
skf.get_n_splits(X, Y)
accuracy=[]

for train_index, test_index in skf.split(X, Y):
    print("TRAIN:", train_index, "TEST:", test_index)
    x_train, x_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = Y.iloc[train_index], Y.iloc[test_index]
    lgbm.fit(x_train,y_train)
    pred=lgbm.predict(x_test)
    score=accuracy_score(pred,y_test)
    accuracy.append(score)
print(accuracy)
```
